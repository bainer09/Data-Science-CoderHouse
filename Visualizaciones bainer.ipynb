```python
"""Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?"""
df_empleados.values.flatten()
fig, ax = plt.subplots(figsize=(12,8))
ax.hist(df_empleados.Salary, bins=(8))
plt.xlim(min(df_empleados.Salary),max(df_empleados.Salary))
ax.set_xlabel('Intervalos de salarios ($)')
ax.set_ylabel('Frecuencia absoluta')
```

```python
"""Realizar un gráfico de violín con los salarios, discriminados por género."""
ax = sns.catplot(data=df_empleados, kind='violin', x='Sex', y='Salary', split=True)
ax.set(xlabel='Genero', ylabel='Salario', title='Salarios distribuidos por genero')

#Decidi utilizar Estado civil como eje x porque la consigna no especificaba la segmentacion del grafico. La mas optima me parecio esta.
```

```python
"""Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)"""
fechas = pd.DatetimeIndex(df_empleados['DateofHire']) #Paso los valores de fecha a el formato correspondiente y guardo en variable
fechas
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(fechas, bins=20) #Utilizo histograma
ax.set_title('Contrato de empleados', fontsize=16)
ax.set_xlabel('Años', fontsize=12)
ax.set_ylabel('Cantidad de empleados contratados',fontsize=12)
plt.xlim(min(fechas), max(fechas))
```
