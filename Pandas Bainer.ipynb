```python
import pandas as pd
import numpy as np
```

```python
df_olimpiada = pd.read_csv('C:/Users/Rodrigo/Downloads/olimpiadas.csv', encoding='latin-1')
df_olimpiada
```

```python
"""1-Construir un objeto de series de tiempo con el indice igual al año de los juegos olimpicos"""
df_olimpiada = df_olimpiada[df_olimpiada['Year'].notna()] #Elimino los valores nan para poder transformar los demas en enteros
fechas_año = df_olimpiada['Year'].astype(int) #Guardo en variable la columna year con formato entero
```

```python
fechas_año_timeindex = pd.DatetimeIndex(fechas_año) #Transformo a indice de tiempo
type(fechas_año_timeindex)
```

```python
columnas = df_olimpiada.columns[2:] #guardo todas las columnas menos la 2 para el nuevo df
valores = df_olimpiada.values[:,2:] #guardo todos los valores menos los de la 2 columna
df_olimpiadas = pd.DataFrame(valores, index=fechas_año_timeindex.year, columns=columnas) #Armo el nuevo df con las variables armadas
df_olimpiadas
```

```python
"""Separar nombres de apellidos en dos columnas distintas usando la coma como separados"""
olimpiadas_nombres = pd.Series(df_olimpiadas['Athlete']) #Separo la columna de atletas
olimpiadas_nombres.str.split(expand=True, n=1) #Divido en dos por nombres y apellidos
```

```python
nombres = olimpiadas_nombres.str.split(expand=True, n=1)
nombres.columns = ['Apellido', 'Nombres'] #Renombro las columnas nuevas
df_olimpiadas = pd.concat([df_olimpiadas, nombres], axis=1) #Agrego al dataframe
df_olimpiadas
```

```python
"""Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?"""
df_olimpiadas.describe() #Medidas Resumen
```

```python
contador_medallas = df_olimpiadas['Country'].value_counts() #Cuento los valores por pais
print('El pais que gano mas medallas es:')
print(contador_medallas[0:1]) #Traigo el primer valor ya que el 'value_counts' me los ordena de mayor a menor
```

```python
"""Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento."""
groups = df_olimpiada.groupby(df_olimpiada.Gender) #Utilizamos el DF inicial
df_h = pd.DataFrame(groups.get_group("Men")) #Elimino los valores de mujeres de la columna 'Gender'
df_h['Year'] = fechas_año #Formateo las fechas 
df_h 
```

```python
h = df_h['Year'].value_counts() #Cuento los valores por año
df_hr = pd.DataFrame(h) #Hago un dataframe con ellos
df_hr
```

```python
df_hr.columns = ["Men's Medals"] #Renombro la columna
df_hr.index.name = 'Year' #Renombro el indice
df_hr.sort_values('Year', ascending=False) #Ordeno por año del mas reciente al menos
```
